```r
library(dplyr)
library(ggplot2)
```

```r
root_dir <- "/data1/APA/Paul_ALS_Data/Mapper_outs_2024/negative_controls/"
sub_cts <- list.files(root_dir, full.names = F)
sub_cts
```

```r
c1_c2s <- list.files(paste0(root_dir,'Astro'), full.names = F)
c1_c2s
```

```r
# make list of sub_cts and c1_c2s for each, so we would have 'root_dor/sub_ct/c1_c2' for each of the sub_ct, but we have multiple c1_c2s for each sub_ct
sub_cts_c1_c2s <- lapply(sub_cts, function(sub_ct){
  lapply(c1_c2s, function(c1_c2){
    paste0(root_dir, sub_ct, '/', c1_c2)
  })
})
sub_cts_c1_c2s
```

```r
# 
pvals_list = c()

# read gene.text in all the sub_cts_c1_c2s dirs, and add the pval column to the pvals_list
for(ct in sub_cts_c1_c2s){
  for(file in ct){
    df <- read.table(paste0(file, '/gene.txt'),sep='\t', header = T)
    df <- df %>% filter(npas >= 2)
    pvals_list <- c(pvals_list, df$pval)

  }}
```

```r
hist(pvals_list, breaks=50, main="Histogram of P-Values", xlab="P-Value", col="blue")
```

```r
# remove NAs from plavs_list
pvals_list <- pvals_list[!is.na(pvals_list)]
```

```r
plot(density(x=pvals_list), main="Density Plot of P-Values", xlab="P-Value", ylab="Density")
```

```r
# Assuming you have a vector of p-values named p_values
library(stats)

# Benjamini-Hochberg FDR correction
p_adjusted_BH <- p.adjust(pvals_list, method = "BH")

# Benjamini-Yekutieli FDR correction (more conservative)
p_adjusted_BY <- p.adjust(pvals_list, method = "BY")
```

```r
# Function to count significant findings for various thresholds
count_significant <- function(p_adjusted, thresholds) {
  sapply(thresholds, function(t) sum(p_adjusted < t))
}

# Define a sequence of potential FDR thresholds
thresholds <- seq(0.000001, 0.01, by=0.001)
# Count significant results for each threshold for both BH and BY adjustments
significant_BH <- count_significant(p_adjusted_BH, thresholds)
significant_BY <- count_significant(p_adjusted_BY, thresholds)

# Plotting the results
plot(thresholds, significant_BH, type = "b", col = "blue", ylim=c(0, max(significant_BH, significant_BY)),
     xlab = "FDR Threshold", ylab = "Number of Significant Findings", main = "Impact of FDR Threshold")
lines(thresholds, significant_BY, type = "b", col = "red")
legend("topright", legend=c("BH", "BY"), col=c("blue", "red"), lty=1, title="Correction Method")
```
